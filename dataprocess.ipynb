```python
import pandas as pd
import numpy as np
import xml.etree.ElementTree as ET
import cv2
import functions as fc
import json
import os
import pickle
```

# training data

```python
files = os.listdir('./dataSet/train_cdc/train_annotations/')
datas = []
for i in files:
    path = './dataSet/train_cdc/train_annotations/' + i
    xml_data = open(path, 'r').read()
    root = ET.XML(xml_data)
    data = []
    for j1, j2 in enumerate(root):
        if j1 == 0:
            fileName = root[0].text
            data.append(fileName)
        
        elif j1 == 1:
            fileFormat = [int(root[1][0].text), int(root[1][1].text), int(root[1][2].text)]
            data.append(fileFormat)
    
        else:
            obj = [root[j1][0].text, int(root[j1][1][0].text),
                   int(root[j1][1][1].text),
                   int(root[j1][1][2].text),
                   int(root[j1][1][3].text)]
            data.append(obj)
    datas.append(data)
```

```python
with open('./dataSet/train_cdc/datas.pickle', 'wb') as f:
    pickle.dump(datas, f)
```

# testing data

```python
path = './dataSet/test_cdc/test_images/'
files = os.listdir('./dataSet/test_cdc/test_images/')
datas = []
for i in files:
    p = path + i
    img = fc.readpic(p)
    x = img.shape
    x = [x[1], x[0]]
    data = [i, x]
    datas.append(data)
```

```python
with open('./dataSet/test_cdc/datas.pickle', 'wb') as f:
    pickle.dump(datas, f)
```

# json

```python
path = './dataSet/train_cdc/train_images/'
Labels = {'aquarium':1
         ,'bottle':2
         ,'bowl':3
         ,'box':4
         ,'bucket':5
         ,'plastic_bag':6
         ,'plate':7
         ,'styrofoam':8
         ,'tire':9
         ,'toilet':10
         ,'tub':11
         ,'washing_machine':12
         ,'water_tower':13
         ,'background':0}
with open('./dataSet/train_cdc/datas.pickle', 'rb') as f:
    datas = pickle.load(f)
    
for i, j in enumerate(datas):
    datas[i][0] = path + datas[i][0]
len(datas)
```

```python
train_images = []
train_objects = []
for i in datas:
    if len(i) == 2:
        continue
    p = i[0]
    train_images.append(p)
    
    objs = i[2:]
    labels = []
    boxes = []
    d = []
    for j in objs:
        labels.append(Labels[j[0]])
        boxes.append(j[1:])
        d.append(False)
    x = {'boxes':boxes, 'labels':labels, 'difficulties':d}
    train_objects.append(x)
```

```python
with open('./dataSet/TRAIN_images.pickle', 'wb') as f:
    pickle.dump(train_images, f)
    
with open('./dataSet/TRAIN_objects.pickle', 'wb') as f:
    pickle.dump(train_objects, f)
```
